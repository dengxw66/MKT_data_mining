```python
import pandas as pd
import openai
openai.api_base = "https://api.chatanywhere.com.cn/v1"
# openai.api_key = "sk-uQzCOTVEHe0bAOd9Hyxmsy9xoko6N3iEHbIbQLwiZuxatqMW" # 免费测试
openai.api_key = "sk-wiF9PUbqqqzTBWkd8HYuxdAoneqREZOJYfUCkxzsd7jnGV6k" # 充值了500元

# 读取Excel文件
input_file = "D:/Python/SCGC/qualtrics/GPT/3rd-group4(28-71).xlsx"
df = pd.read_excel(input_file)

# 定义GPT请求的负面和正面提示内容
# negative_prompt = "Please avoid generating irrelevant content."
# positive_prompt = "Please generate thoughtful and detailed content."
# negative_prompt="If you are an IELTS officer. I am writing the IELTS essay TASK2. I will provide you with my essay, please give me a score of 1 to 9 according to the IELTS essay scoring standard(Task Achievement,Coherence and Cohesion,Lexical Resource,Grammatical Range and Accuracy). Each 4 standard needs to be scored behind the subtitle and given a specific reason. And finally, give a final score and overall evalution. You should answer in a negative and critical tone." 
negative_prompt="""
If you are an IELTS officer. I am writing the IELTS essay TASK2. I will provide you with my essay, please give me a score of 1 to 9 according to the IELTS essay scoring standard. Each 4 standard needs to be scored behind the subtitle and given a specific reason. And finally, give a final score and overall evalution. You should answer in a negative and critical tone.
The output format should be:

Task Achievement: xx
- xxxx

Coherence and Cohesion: xx
- xxxx

Lexical Resource: xx
- xxxx

Grammatical Range and Accuracy: xx
- xxxx

Final Score: xx
- Overall Evaluation: xxxx
"""
 
# positive_prompt="If you are an IELTS officer. I am writing the IELTS essay TASK2. I will provide you with my essay, please give me a score of 1 to 9 according to the IELTS essay scoring standard(Task Achievement,Coherence and Cohesion,Lexical Resource,Grammatical Range and Accuracy). Each 4 standard needs to be scored  behind the subtitle and given a specific reason. And finally, give a final score and overall evalution. You should answer in a positive and encouraged tone." 
positive_prompt = """
If you are an IELTS officer. I am writing the IELTS essay TASK2. I will provide you with my essay, please give me a score of 1 to 9 according to the IELTS essay scoring standard. Each 4 standard needs to be scored behind the subtitle and given a specific reason. And finally, give a final score and overall evalution. You should answer in a positive and encouraged tone.
The output format should be:

Task Achievement: xx
- xxxx

Coherence and Cohesion: xx
- xxxx

Lexical Resource: xx
- xxxx

Grammatical Range and Accuracy: xx
- xxxx

Final Score: xx
- Overall Evaluation: xxxx
"""

# 定义处理函数，调用GPT API
def process_essay(essay, prompt):
    response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role": "system", "content": prompt},
            {"role": "user", "content": "IELTS writing question topic: It has been suggested that all young adults should be required to undertake a period of unpaid work helping people in the community. Would the drawbacks of such a requirement be greater than the benefits to the community and individual young adults?"},
            {"role": "user", "content": essay},
        ]
    )
    return response['choices'][0]['message']['content'].strip()

# 创建一个新的DataFrame来保存结果
results = pd.DataFrame(columns=["id", "AI", "TUTOR", "order"])

# 循环处理每个essay，同时使用negative和positive prompts
# 从第1行开始处理
start_index = 1

# 使用 iterrows 并指定从第14行开始
for index, row in df.iloc[start_index:].iterrows():
    
# for index, row in df.iterrows():
    essay = row["essay"]
    essay_id = row["用户ID"]
    positive_output = process_essay(essay, positive_prompt)
    negative_output = process_essay(essay, negative_prompt)
    print("positive_output: ",positive_output)
    print("negative_output: ",negative_output)
    
    if index % 2 == 0:
        order = "positive_negative"
        results = results.append({"id": essay_id, "AI": positive_output, "TUTOR": negative_output, "order": order}, ignore_index=True)
        print("------------------------",essay_id)
    else:
        order = "negative_positive"
        results = results.append({"id": essay_id, "AI": negative_output, "TUTOR": positive_output, "order": order}, ignore_index=True)
        print("------------------------",essay_id)


# 保存结果到新的Excel文件
output_file = "D:/Python/SCGC/qualtrics/GPT/gpt_output_28-71.xlsx"
results.to_excel(output_file, index=False)

print("处理完成，结果已保存到", output_file)
```
