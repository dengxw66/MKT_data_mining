```python
# 返回单词保存

import pandas as pd
import jieba
import string
import re
import emoji

# Function to load stopwords
def load_stopwords(stopwords_file_path):
    with open(stopwords_file_path, 'r', encoding='utf-8') as f:
        stopwords = set(f.read().splitlines())
    return stopwords

# Function to remove emojis
def remove_emojis(text):
    return emoji.replace_emoji(text, replace='')

# Function to clean text
def clean_text(text, stopwords):
    # Remove emojis
    text = remove_emojis(text)
    # Remove English letters
    text = re.sub(r'[A-Za-z]', '', text)
    # Remove punctuation and special characters
    text = re.sub(r'[{}]+'.format(string.punctuation), '', text)
    # Remove digits
    text = re.sub(r'\d+', '', text)
    # Remove extra whitespace
    text = re.sub(r'\s+', ' ', text).strip()
    # Tokenize and remove stopwords
    words = jieba.lcut(text)
    words = [word for word in words if word not in stopwords and word.strip()]
    return ' '.join(words)

# Load data
# excel_file_path = "/data1/dxw_data/llm/RAG-mkt-kmeans/data1/raw_content_dianping_hotpot.xlsx"
# excel_file_path = "/data1/dxw_data/llm/RAG-mkt-kmeans/data2/raw_content_dianping_barbecue.xlsx"
excel_file_path = "/data1/dxw_data/llm/RAG-mkt-kmeans/data3/raw_content_dianping_cake.xlsx"
df = pd.read_excel(excel_file_path)

# Load stopwords
stopwords_file_path = "/data1/dxw_data/llm/ML/LIWC/datasets/stopwords_cn.txt"
stopwords = load_stopwords(stopwords_file_path)

# Clean the "comment" column
df['cleaned_comment'] = df['comment'].apply(lambda x: clean_text(str(x), stopwords))

# Save the cleaned data to a new Excel file
# output_file_path = "/data1/dxw_data/llm/RAG-mkt-kmeans/data1/cleaned_comments_dianping_hotpot.xlsx"
# output_file_path = "/data1/dxw_data/llm/RAG-mkt-kmeans/data2/cleaned_comments_dianping_barbecue.xlsx"
output_file_path = "/data1/dxw_data/llm/RAG-mkt-kmeans/data3/cleaned_comments_dianping_cake.xlsx"
df.to_excel(output_file_path, index=False)

print("Data cleaning complete. Cleaned data saved to:", output_file_path)
```
