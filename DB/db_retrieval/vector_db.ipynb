```python

# import
from langchain.embeddings.sentence_transformer import SentenceTransformerEmbeddings
from langchain.text_splitter import CharacterTextSplitter
from langchain.vectorstores import Chroma
from langchain.document_loaders import TextLoader
from langchain.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
```

```python
# load the document and split it into chunks
loader = TextLoader("/data1/dxw_data/llm/RA/Twitter.txt", encoding="utf-8")
documents = loader.load()
```

```python
# split it into chunks
text_splitter = CharacterTextSplitter(chunk_size=100, chunk_overlap=0)
docs = text_splitter.split_documents(documents)

# create the open-source embedding function
model_kwargs = {'device': 'cuda:7'}
embedding_function = HuggingFaceEmbeddings(model_name='/data1/dxw_data/llm/text2vec-large-chinese',model_kwargs=model_kwargs) # ‰ºöÊä•Èîô‚ÄúNo sentence-transformers model found‚Äù‰ΩÜÊòØ‰∏çÂΩ±Âìç‰ΩøÁî®,ËøôÂè™ÊòØhuggingfaceÁöÑÊ£ÄÊµãÈóÆÈ¢ò„ÄÇ


# load it into Chroma
# db = Chroma.from_documents(docs, embedding_function)
db = FAISS.from_documents(docs, embedding_function)

```

```python
# retriever = db.as_retriever()
# retriever = db.as_retriever(search_type="mmr")
retriever = db.as_retriever(search_kwargs={"k": 15}) # k chontrol the number of feedback
```

```python
docs = retriever.invoke("multimodal,image,text")
print(len(docs))

# print results
for doc in docs:
    print("="*100)
    print(doc.page_content)
```
