```python
# 使用chrome开发者模式
option = webdriver.ChromeOptions()
# 屏蔽webdriver特征
option.add_experimental_option(
    "excludeSwitches", ["enable-automation"]
)  
# 禁用启用Blink运行时的功能，关闭网页页面的反扒校验
option.add_argument("--disable-blink-features=AutomationControlled")

option.add_argument("--no-sandbox") # 表示禁用沙盒模式
# 沙盒模式：每一个标签页都是一个沙盒（sandbox），以“防止恶意软体自行执行安装”或“利用一个分页影响其他的分页”
# 关闭web沙盒的命令，有可能导致浏览恶意网站时，被入侵。 
option.add_argument("--disable-dev-shm-usage") # 启动浏览器时，添加参数--disable-dev-shm-usage。该参数使用本地local/tmp代替/dev/shm作为 Chrome 的运行空间，local/tmp比/dev/shm有更大的空间，可以使Cypress运行时，不容易因为一个文件的测试用例数多，导致内存溢出的问题。

# 指定chromedriver的路径
chrome_service = Service(r'C:/Users/89859/AppData/Local/Google/Chrome/Application/chromedriver.exe')
# 创建Chrome WebDriver对象
driver = webdriver.Chrome(service=chrome_service, options=option)
# Selenium执行cdp命令，没有出现验证框就能登录说明已经成功屏蔽网页对selenium的识别。

driver.execute_cdp_cmd("Page.addScriptToEvaluateOnNewDocument", {
    "source": """
                Object.defineProperty(navigator, 'webdriver', {
                    get: () => undefined
                })
                """
})

driver.set_page_load_timeout(25) # 用于设置页面加载超时,指定时间内未加载出页面则会报错
driver.delete_all_cookies()  # 删除所有 cookie 信息
url = (
    "https://www.mijian360.com/group/fa/" 
)
driver.get(url)
time.sleep(1)    # 使用chrome开发者模式
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, NoSuchElementException
import time

# driver = webdriver.Chrome()
driver.get("https://www.mijian360.com/group/fa/")
time.sleep(5)

total_pages = 50
elements_per_page = 10

contents = []

def find_element_by_xpaths(driver, xpaths, timeout=5):
    for xpath in xpaths:
        try:
            element = WebDriverWait(driver, timeout).until(
                EC.presence_of_element_located((By.XPATH, xpath))
            )
            return element.text,xpath
        except TimeoutException:
            print(f"Timeout while trying to find element with XPath: {xpath}")
        except NoSuchElementException:
            print(f"No such element found with XPath: {xpath}")
        except Exception as e:
            print(f"Unknown error occurred: {e}")
    return None

def get_element_text(driver, xpath, timeout=5):
    try:
        element = WebDriverWait(driver, timeout).until(
            EC.presence_of_element_located((By.XPATH, xpath))
        )
        return element.text
    except TimeoutException:
        print(f"Timeout while trying to get element with XPath: {xpath}")
        return None
    except NoSuchElementException:
        print(f"No such element found with XPath: {xpath}")
        return None
    except Exception as e:
        print(f"Unknown error occurred: {e}")
        return None

for page in range(total_pages):
    for i in range(1, elements_per_page + 1):
        try:
            time.sleep(5)
            view_count_xpaths = [f'/html/body/div/div/div/div[2]/div[2]/div/div[1]/div[1]/div/div[3]/div[2]/div[{i}]/a/div/div[3]/div[1]/span',
                         f'/html/body/div/div/div/div[2]/div[2]/div/div[1]/div[1]/div/div[3]/div[2]/a[{i}]/div/div[3]/div[1]/span']
            title_xpaths = [f'/html/body/div/div/div/div[2]/div[2]/div/div[1]/div[1]/div/div[3]/div[2]/div[{i}]/a/div/div[1]',
                            f'/html/body/div/div/div/div[2]/div[2]/div/div[1]/div[1]/div/div[3]/div[2]/a[{i}]/div/div[1]']
            time_xpaths = [f'/html/body/div/div/div/div[2]/div[2]/div/div[1]/div[1]/div/div[3]/div[2]/div[{i}]/a/div/div[3]/div[2]',
                   f'/html/body/div/div/div/div[2]/div[2]/div/div[1]/div[1]/div/div[3]/div[2]/a[{i}]/div/div[3]/div[2]']

            # content_xpath = '/html/body/div/div/div/div[2]/div[2]/div/div[1]/div[1]/div/div[1]/div[2]/div'
            content_xpath ="/html/body/div/div/div/div[2]/div[2]/div/div[1]/div[1]/div/div[1]/div[2]/div"
            # keywords_xpath = '/html/body/div/div/div/div[2]/div[2]/div/div[1]/div[1]/div/div[1]/div[1]/div/div/div[2]/div'

            view_count,view_countxpath = find_element_by_xpaths(driver, view_count_xpaths)
            title,titlexpath = find_element_by_xpaths(driver, title_xpaths)
            time_posted,time_postedxpath = find_element_by_xpaths(driver, time_xpaths)
            
            # view_count = get_element_text(driver, view_count_element)
            print("view_count: ",view_count)
            # title = get_element_text(driver, title_element)
            print("title: ",title)
            # time_posted = get_element_text(driver, time_element)
            print("time_posted: ",time_posted)

            if not view_count or not title or not time_posted:
                continue

            main_window_handle = driver.current_window_handle
            driver.find_element(By.XPATH, view_countxpath).click()
            WebDriverWait(driver, 5).until(EC.new_window_is_opened)
            
            for handle in driver.window_handles:
                if handle != main_window_handle:
                    driver.switch_to.window(handle)
                    break

            content = get_element_text(driver, content_xpath)
            print("content: ",content)
            # keywords = get_element_text(driver, keywords_xpath)
            # print("keywords: ",keywords)
            contents.append([view_count, title, time_posted, content])
            print(f"-----------------已经得到第{page * elements_per_page + i}个元素的数据-----------------")
            
            driver.close()
            driver.switch_to.window(main_window_handle)

        except Exception as e:
            print(f"获取第{page * elements_per_page + i}个元素的数据时发生未知错误: {e}")
            contents.append([view_count, title, time_posted, ""])
            print(f"-----------------已经得到第{page * elements_per_page + i}个元素的数据-----------------")

    next_page_xpath = '/html/body/div/div/div/div[2]/div[2]/div/div[1]/div[1]/div/div[3]/div[4]/div/ul/li[13]/span'
    try:
        next_page_element = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, next_page_xpath))
        )
        next_page_element.click()
        time.sleep(5)
    except Exception as e:
        print(f"翻页时发生错误: {e}")
        break

print("-----------------数据获取完成-----------------")
for content in contents:
    print("hotness: ", content[0])
    print("title: ", content[1])
    print("time: ", content[2])
    print("content: ", content[3])

# 将数据保存到Excel

df = pd.DataFrame(contents, columns=['hotness', 'title', 'time', 'content'])
df.to_excel('D:\Python\SCGC\medical-platform\data_肺癌.xlsx', index=False)
print("数据已保存到data.xlsx")
```
